```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, precision_score, recall_score
from sklearn.preprocessing import LabelEncoder


# create label encoder object
le = LabelEncoder()

# encode categorical variables
df['Age'] = le.fit_transform(df['Age'])
df['Ease in Booking'] = le.fit_transform(df['Ease in Booking'])
df['Travel Time'] = le.fit_transform(df['Travel Time'])
df['Booking Time'] = le.fit_transform(df['Booking Time'])
df['Class'] = le.fit_transform(df['Class'])
df['Air Route'] = le.fit_transform(df['Air Route'])
df['Price Range (INR)'] = le.fit_transform(df['Price Range (INR)'])
df['Pricing Satisfaction'] = le.fit_transform(df['Pricing Satisfaction'])
df['Check-in Experience'] = le.fit_transform(df['Check-in Experience'])
df['Communication Satisfaction'] = le.fit_transform(df['Communication Satisfaction'])
df['Flight Delayed'] = le.fit_transform(df['Flight Delayed'])
df['Seat Comfort'] = le.fit_transform(df['Seat Comfort'])
df['Cleanliness Satisfaction'] = le.fit_transform(df['Cleanliness Satisfaction'])
df['Food Quality Rating'] = le.fit_transform(df['Food Quality Rating'])
df['Entertainment Satisfaction'] = le.fit_transform(df['Entertainment Satisfaction'])
df['Attendant Friendliness Rating'] = le.fit_transform(df['Attendant Friendliness Rating'])
df['Baggage Claim Satisfaction'] = le.fit_transform(df['Baggage Claim Satisfaction'])
df['Immigration Efficiency Rating'] = le.fit_transform(df['Immigration Efficiency Rating'])
df['Ground Transportation Satisfaction'] = le.fit_transform(df['Ground Transportation Satisfaction'])
df['Airline'] = le.fit_transform(df['Airline'])
df['Overall Satisfaction'] = le.fit_transform(df['Overall Satisfaction'])


```

```python
df.head()
```

```python

# Split the dataset into training and testing sets
X = df.drop(['Overall Satisfaction'], axis=1)
y = df['Overall Satisfaction']
X_train_dtc, X_test_dtc, y_train_dtc, y_test_dtc = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the decision tree classifier with a max depth of 5
DT_clf = DecisionTreeClassifier(max_depth=5, random_state=42)

# Train the classifier on the training data
DT_clf.fit(X_train_dtc, y_train_dtc)

#Make predictions on the training and testing data
y_pred_train_dtc = DT_clf.predict(X_train_dtc)
y_pred_test_dtc = DT_clf.predict(X_test_dtc)

#Evaluate the performance of the classifier on training data

train_acc_score = accuracy_score(y_train_dtc, y_pred_train_dtc)
train_conf_mat_dtc = confusion_matrix(y_train_dtc, y_pred_train_dtc)

print('Decision Tree Training Accuracy:', train_acc_score)
print('Decision Tree Training Confusion Matrix:')
print(train_conf_mat_dtc)

#Evaluate the performance of the classifier on testing data

test_acc_score = accuracy_score(y_test_dtc, y_pred_test_dtc)
test_conf_mat_dtc = confusion_matrix(y_test_dtc, y_pred_test_dtc)

print('Decision Tree Testing Accuracy:', test_acc_score)
print('Decision Tree Testing Confusion Matrix:')
print(test_conf_mat_dtc)
```

```python
import numpy as np

# Get the feature importance scores
importances = DT_clf.feature_importances_

# Print the feature importance scores in descending order
indices = np.argsort(importances)[::-1]

names = [X_train_dtc.columns[i] for i in indices]

# Plot feature importance
plt.figure(figsize=(10,6))
plt.title("Feature Importance")
plt.bar(range(X_train_dtc.shape[1]), importances[indices])
plt.xticks(range(X_train_dtc.shape[1]), names, rotation=90)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix

# Split the data into training and testing sets
X_train_rfc, X_test_rfc, y_train_rfc, y_test_rfc = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the Random Forest classifier with hyperparameters to tune
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Define the hyperparameters to tune
params = {
    'max_depth': [2, 4, 6, 8],
    'min_samples_split': [2, 4, 6],
    'min_samples_leaf': [1, 2, 4]
}

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(rf_clf, params, cv=5, n_jobs=-1)
grid_search.fit(X_train_rfc, y_train_rfc)


# Print the best hyperparameters found
print('Best hyperparameters:', grid_search.best_params_)

# Make predictions on the testing data
y_pred_rfc = grid_search.predict(X_test_rfc)

# Evaluate the performance of the classifier on the testing set
accuracy_test = accuracy_score(y_test_rfc, y_pred_rfc)
confusion_mat_test = confusion_matrix(y_test_rfc, y_pred_rfc)

# Evaluate the performance of the classifier on the training set
y_pred_train_rfc = grid_search.predict(X_train_rfc)
accuracy_train = accuracy_score(y_train_rfc, y_pred_train_rfc)
confusion_mat_train = confusion_matrix(y_train_rfc, y_pred_train_rfc)


print('Random Forest Classifier (with GridSearchCV) Training Accuracy:', accuracy_train)
print('Random Forest Classifier (with GridSearchCV) Training Confusion Matrix:')
print(confusion_mat_train)

print('Random Forest Classifier (with GridSearchCV) Test Accuracy:', accuracy_test)
print('Random Forest Classifier (with GridSearchCV) Test Confusion Matrix:')
print(confusion_mat_test)

```

```python
# Get feature importances from the trained classifier
importances = grid_search.best_estimator_.feature_importances_

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [X.columns[i] for i in indices]

# Create plot
plt.figure()

# Create plot title
plt.title("Feature Importance")

# Add bars
plt.bar(range(X.shape[1]), importances[indices])

# Add feature names as x-axis labels
plt.xticks(range(X.shape[1]), names, rotation=90)

# Show plot
plt.show()
```

```python
# Use the best estimator to make predictions on the train and test datasets
y_train_pred_rfc = grid_search.best_estimator_.predict(X_train_rfc)
y_test_pred_rfc = grid_search.best_estimator_.predict(X_test_rfc)

# Print the accuracy on the train and test datasets
print('Train accuracy:', accuracy_score(y_train_rfc, y_train_pred_rfc))
print('Test accuracy:', accuracy_score(y_test_rfc, y_test_pred_rfc))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# create a logistic regression object
logreg = LogisticRegression(max_iter=10000)

X = df.drop(['Overall Satisfaction'], axis=1)
y = df['Overall Satisfaction']

# Split the data into training and testing sets
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(X, y, test_size=0.3, random_state=42)


# define the hyperparameters to tune
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}

# create a GridSearchCV object
grid_search = GridSearchCV(logreg, param_grid, cv=5)

# fit the GridSearchCV object to the data
grid_search.fit(X_train_lr, y_train_lr)

# print the best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# evaluate the model on the training set
y_pred_train_lr = grid_search.predict(X_train_lr)
train_acc = accuracy_score(y_train_lr, y_pred_train_lr)
print("Train accuracy:", train_acc)

# evaluate the model on the test set
y_pred_test_lr = grid_search.predict(X_test_lr)
test_acc = accuracy_score(y_test_lr, y_pred_test_lr)
print("Test accuracy:", test_acc)

```
